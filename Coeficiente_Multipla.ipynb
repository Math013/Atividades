```python
def quadr(x):
  Q = []
  for e in x:
    Q.append(e**2)
  return Q

y = []
x1 = []
x2 = []

n = int(input("Informe o número de observações: ")) # Tamanho de n
print()

for e in range(0, n):
  Y = int(input(f"Informe o {e+1}º valor em y: ")) # Valores de X
  y.append(Y)

sy = sum(y) # Soma dos elementos X
print()
print(f'Elementos de y: {y}')
print(f'A soma de y é: {sy}') # Soma dos elementos em X
print()

# =====================================================

for e in range(0, n):
  x_1 = int(input(f"Informe o {e+1}º valor em x1: "))
  x1.append(x_1)

sx1 = sum(x1)
print()
print(f'Elementos de x1: {x1}')
print(f'A soma de x1 é: {sx1}') # Soma dos elementos em X
print()

# ======================================================

for e in range(0, n):
  x_2 = int(input(f"Informe o {e+1}º valor em x2: "))
  x2.append(x_2)

sx2= sum(x2)
print()
print(f'Elementos de x2: {x2}')
print(f'A soma de x2 é: {sx2}') # Soma dos elementos em X
print()

# ======================================================

med_x1 = sx1 / n # Média de x1 (Variável auxiliar)
med_x2 = sx2 / n # Média de x2 (Variável auxiliar)
med_y = sy / n # Média de y (Variável auxiliar)

print(f'A média de x1 é {med_x1}')
print(f'A média de x2 é {med_x2}')
print(f'A média de y é {med_y}')

print()

# med_x1 -> x1med -> X1
# med_x2 -> x2med -> X2
# med_y -> ymed -> Y

x1med = [] # Variável auxiliar de X1, essa variável recebe n vezes a média de x1, logo auxilia na inserção de valores na lista X1 (X1 = x1 - x1med)
x2med = [] # Variável auxiliar de X2, essa variável recebe n vezes a média de x1, logo auxilia na inserção de valores na lista X2 (X2 = x2 - x2med)
ymed = [] # Variável auxiliar de Y, essa variável recebe n vezes a média de Y, logo auxilia na inserção de valores na lista Y (Y = y - ymed)
cont = 0 

while cont < n: # Gambiarra (Inserir Média de y em uma lista para poder subtrair com os elementos de y)
  ymed.append(med_y) # Insere a média de y n vezes na lista ymed, que irá ser utilizada na lista Y
  x1med.append(med_x1) # Insere a média de x1, n vezes na lista x1med, que irá ser utilizada na lista X1
  x2med.append(med_x2) # Insere a média de x2 n vezes na lista x2med, que irá ser utilizada na lista X2
  cont+= 1

Y = [ex-ey for ex,ey in zip(y, ymed)] # Y = y - ymed

if sum(Y) != 0: # A Soma de Y não pode ser 0
  print("A soma de Y está errada!")
  print("Elementos errados!")

X1 = [ex-ey for ex,ey in zip(x1, x1med)] # X1 = x1 - x1med

if sum(X1) != 0:
  print("A soma de X1 está errada!") # A Soma de X1 não pode ser 0
  print("Elementos errados!")

X2 = [ex-ey for ex,ey in zip(x2, x2med)] # X2 = x2 - x2med

if sum(X2) != 0:
  print("A soma de X2 está errada!") # A Soma de X2 não pode ser 0
  print("Elementos errados!")

# ===========================================

print(f'Elementos de Y: {Y}')
print(f"Soma de Y: {sum(Y)}") #Opcional
print("OBS: A soma TEM que dar 0!")
print()

print(f'Elementos de X1: {X1}')
print(f"Soma de X1: {sum(X1)}") # Opcional
print("OBS: A soma TEM que dar 0!") 
print()

print(f'Elementos de X2: {X2}')
print(f"Soma de X2: {sum(X2)}") # Opcional
print("OBS: A soma TEM que dar 0!")
print()

# ============================================

YX1 = [ely*elx1 for ely,elx1 in zip(Y, X1)] # Y * X1
SYX1 = sum(YX1) # Variável auxiliar (Soma de Y*X1)
print(f'Elementos de YX1: {YX1}')
print(f'Soma de YX1: {SYX1}')
print()

YX2 = [ely*elx2 for ely,elx2 in zip(Y, X2)] # Y * X2
SYX2 = sum(YX2) # Variével auxiliar (Soma de Y*X2)
print(f'Elementos de YX2: {YX2}')
print(f'Soma de YX2: {SYX2}')
print()

X1X2 = [x1_* x2_ for x1_,x2_ in zip(X1, X2)] # X1 * X2
SX1X2 = sum(X1X2) # Variável auxiliar (Soma de X1*X2)
print(f'Elementos de X1X2: {X1X2}')
print(f'Soma de X1X2: {SX1X2}')
print()

Y_Quadr = quadr(Y) # Y²
SY_Quadr = sum(Y_Quadr) # Variável auxiliar (Soma de Y²)
print(f"Elementos Y²: {Y_Quadr}")
print(f'Soma de Y²: {SY_Quadr}')
print()

X1_Quadr = quadr(X1) # X1²
SX1_Quadr = sum(X1_Quadr) # Variável auxiliar (Soma de X1²)
print(f"Elementos X1²: {X1_Quadr}")
print(f'Soma de X1²: {SX1_Quadr}')
print()

X2_Quadr = quadr(X2) # X2²
SX2_Quadr = sum(X2_Quadr) # Variável auxiliar (Soma de X2²)
print(f"Elementos X2²: {X2_Quadr}")
print(f"Soma de X2²: {SX2_Quadr}")
print()

# =============================================================================

B1_part1 = (SYX1 * SX2_Quadr) - (SYX2 * SX1X2) # Parte 1 da Fórmula de B1
B1_part2 = (SX1_Quadr * SX2_Quadr) - (SX1X2 ** 2) # Parte 2 da Fórmula de B1
B1 = B1_part1 / B1_part2 # Fórmula de B1

print(f'B1 = {SYX1} * {SX2_Quadr} - {SYX2} * {SX1X2} / {SX1_Quadr} * {SX2_Quadr} - ({SX1X2})²')
print(f'O resultado de B1: {B1:.5f}')
print()

# =============================================================================

B2_part1 = (SYX2 * SX1_Quadr) - (SYX1 * SX1X2) # Parte 1 da Fórmula de B24
B2_part2 = (SX1_Quadr * SX2_Quadr) - (SX1X2 ** 2)
B2 = B2_part1 / B2_part2

print(f'B2 = {SYX2} * {SX1_Quadr} - {SYX1} * {SX1X2} / {SX1_Quadr} * {SX2_Quadr} - ({SX1X2})²')
print(f'O resultado de B2: {B2:.5f}')
print()

# =============================================================================

B0 = med_y - (B1 * med_x1) - (B2 * med_x2)
print(f'B0 = {med_y} - {B1:.5f} * {med_x1} - {B2:.5f} * {med_x2}')
print(f'O resultado de B0: {B0:.5f}')

print()
print(f'A fórmula final é: Ŷ = {B0:.2f} + {B1:.2f}x1 + {B2:.2f}x2')
```
